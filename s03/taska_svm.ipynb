```python
from sklearn import svm, metrics
from PIL import Image
import numpy as np
import os
from sklearn.model_selection import train_test_split
```

```python
images = []
for r,d,_ in os.walk('./mnist/train') :
    if d :
        for directory in d:
            for r2,d2,f2  in os.walk(r + '/' + directory):
                for img_name in f2 :
                    img_path = r + '/' + directory + '/' + img_name
                    images.append((img_path,directory))
```

```python
x_data = np.array([np.asarray(Image.open(img_path).convert('L')) for img_path,_ in images])
y_data = np.array([int(label) for _, label in images])
x_data = x_data.reshape(len(x_data),28 * 28) / 255.0
```

```python
from matplotlib.colors import Normalize
import matplotlib.pyplot as plt

class MidpointNormalize(Normalize):

    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))

def plot_param_space_heatmap(scores, C_range, gamma_range, name = ""):
    plt.figure(figsize=(8, 6))
    plt.subplots_adjust(left=.2, right=0.95, bottom=0.15, top=0.95)
    plt.imshow(scores, interpolation='nearest', cmap=plt.cm.jet,
               norm=MidpointNormalize(vmin=0.5, midpoint=0.9))
    plt.xlabel('gamma')
    plt.ylabel('C')
    plt.colorbar()
    plt.xticks(np.arange(len(gamma_range)), gamma_range, rotation=45)
    plt.yticks(np.arange(len(C_range)), C_range)
    plt.title('Validation accuracy for ' + name)
    plt.show()
```

```python
# GridSearch
from sklearn.model_selection import GridSearchCV

rand_idx_GS = np.random.choice(len(x_data),3000)
x_data_GS = x_data[rand_idx_GS]
y_data_GS = y_data[rand_idx_GS]

x_train, x_test, y_train, y_test = train_test_split(
    x_data_GS,
    y_data_GS,
    test_size=0.2,
    random_state=42)

C_params = np.outer(np.logspace(-1, 1, 3),np.array([1,5]))
C_params = C_params.flatten()

gamma_params = np.outer(np.logspace(-3, 0, 4),np.array([1,5]))
gamma_params = gamma_params.flatten()

kernel_params = ['rbf','poly']

grid_classifier = GridSearchCV(
    estimator = svm.SVC(),
    param_grid = {
        'kernel' : kernel_params,
        'C' : C_params,
        'gamma' : gamma_params
    },
    n_jobs = -1,
    verbose = 2
)

grid_classifier.fit(x_train,y_train)

sorted(grid_classifier.cv_results_.keys())

classifier = grid_classifier.best_estimator_
params = grid_classifier.best_params_

for k in kernel_params :
    scores = grid_classifier.cv_results_['mean_test_score'][grid_classifier.cv_results_['param_kernel'] == k].reshape(
        len(C_params),
        len(gamma_params)
    )
    plot_param_space_heatmap(scores, C_params, gamma_params, name = k)
```

```python
import operator
print(grid_classifier.best_params_)
print()
means = grid_classifier.cv_results_['mean_test_score']
stds = grid_classifier.cv_results_['std_test_score']
for mean, std, params in sorted(zip(means, stds, grid_classifier.cv_results_['params']),reverse=True, key=operator.itemgetter(0)):
    print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))
print()
```

```python
# rand_idx= np.random.choice(len(x_data),200)
# x_data = x_data[rand_idx]
# y_data = y_data[rand_idx]

x_train, x_test, y_train, y_test = train_test_split(
    x_data,
    y_data,
    test_size=0.2,
    random_state=42)

classifier = svm.SVC(kernel = grid_classifier.best_params_['kernel'],gamma=grid_classifier.best_params_['gamma'], C = grid_classifier.best_params_['C'])
classifier.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.figure(1, figsize=(15, 12), dpi=160)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()  

expected = y_test
predicted = classifier.predict(x_test)

print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(expected, predicted)))
      
cm = metrics.confusion_matrix(expected, predicted)
print("Confusion matrix:\n%s" % cm)

plot_confusion_matrix(cm)

print("Accuracy={}".format(metrics.accuracy_score(expected, predicted)))
```

```python
rand_idx= np.random.choice(len(x_data),len(x_data))
x_data_cv = x_data[rand_idx]
y_data_cv = y_data[rand_idx]

# cross validation
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit

classifier = svm.SVC(
    kernel = grid_classifier.best_params_['kernel'],
    gamma  = grid_classifier.best_params_['gamma'],
    C      = grid_classifier.best_params_['C']
)
cv = ShuffleSplit(n_splits = 10, test_size = 0.3, random_state = 0)
scores = cross_val_score(classifier, x_data_cv, y_data_cv, cv=cv)

print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
